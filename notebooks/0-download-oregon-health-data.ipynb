```python
import datetime
import requests
from pathlib import Path
```

```python
pdf_folder = Path('../data/oregon/')
pdf_folder.mkdir(parents=True, exist_ok=True)

xml_folder = Path('../data/oregon-xml/')
xml_folder.mkdir(parents=True, exist_ok=True)
```

```python
#CHANGE ACCORDINGLY
report_date = datetime.date(2021, 12, 1)

while report_date > datetime.date(2021, 3, 10):
    url = f'https://www.oregon.gov/oha/covid19/Documents/DataReports/Weekly-Outbreak-COVID-19-Report-{report_date}-FINAL.pdf'
    r = requests.get(url)
    status_check = r.status_code
    
    # If it doesn't work, try a different url format
    if status_check == 404:
        modified_date = report_date.strftime("%Y-%-m-%-d") 
        url = f'https://www.oregon.gov/oha/covid19/Documents/DataReports/COVID-19-Weekly-Outbreak-Report-{modified_date}-FINAL.pdf'
        r = requests.get(url)
    
    if r.status_code != 404:
        filename = Path(pdf_folder, url.split("/")[-1])
        filename.write_bytes(r.content)
        print(f"saved {report_date}")
    else:
        print(f'had issues with {report_date}')
    
    report_date = report_date - datetime.timedelta(days=7)
```

```python
leftovers = ['https://www.oregon.gov/oha/covid19/Documents/DataReports/Weekly-Outbreak-COVID-19-Report-2021-09-09-FINAL.pdf',
 'https://www.oregon.gov/oha/covid19/Documents/DataReports/Weekly-Outbreak-COVID-19-2021-10-06-FINAL.pdf']

for url in leftovers:
    r = requests.get(url)
    if r.status_code != 404:
        filename = Path(pdf_folder, url.split("/")[-1])
        filename.write_bytes(r.content)
```
